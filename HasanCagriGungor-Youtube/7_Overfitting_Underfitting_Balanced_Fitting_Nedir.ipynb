```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("datasets/diamonds")
df.head(5)
```

```python
df = pd.get_dummies(df, columns=["cut", "color", "clarity"], drop_first=True)
df.head(5)
```

```python
y = df[["price"]]
x = df.drop("price", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=19, train_size=0.76)
```

```python
lm = LinearRegression()
model = lm.fit(x_train, y_train)
```

```python
test_score = model.score(x_test, y_test)
```

```python
train_score = model.score(x_train, y_train)
```

```python
# test_score ile train_score birbirine ne kadar yakınsa o kadar dengeli öğrenmedir.     BALANCED FITTING
# test_score ile train_score birbirinden ne kadar uzaksa o kadar aşırı öğrenmedir.      OVER-UNDER FITTING
```
