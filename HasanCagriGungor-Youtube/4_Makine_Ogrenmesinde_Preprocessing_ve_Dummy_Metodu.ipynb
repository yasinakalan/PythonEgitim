```python
# VERİ ÖN İŞLEME --> DATA PREPROCESSING
# PANDAS DUMMY METODU --> STRING İFADELERİ SAYISAL VERİLERE ÇEVİRME

# Video: https://www.youtube.com/watch?v=8enRun3anB8&list=PL3ED48mWmYxrAdWjQlOWzFNaM4gLgry5T&index=4
```

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/Audi_A1_listings.csv")
df.head(5)
```

```python
# y = a*x + b
```

```python
# GEREKSİZ VERİLERİN DATAFRAME DEN ATILMASI:

df = df.drop(columns=["index", "href", "MileageRank", "PriceRank", "PPYRank", "Score"])        # Fiyata etkisi olmayan sütunlar verisetinden çıkarılmalıdır.
df.head(5)
```

```python
# SÜTUN İSİMLERİNİN DEĞİŞTİRİLMESİ:
df.columns=["yil", "kasa", "mil", "motor", "ps", "vites", "yakit", "sahip", "fiyat", "ppy"]
df.head(5)
```

```python
# STRING VERİLERİN TAMAMEN SAYISAL OLARAK DEĞİŞTİRİLMESİ:
df.info()
```

```python
df["motor"] = df["motor"].str.replace("L","")       # motor sütunundaki verilerden "L" ifadesi hiçbirşey("") ile değiştirildi.
```

```python
df["motor"] = pd.to_numeric(df["motor"])            # "L" atıldıktan sonra sayılar hala string olduğundan bu aşamada float veri tipine dönüştürüldü.
df.head(5)
```

```python
df = pd.get_dummies(df, columns=["kasa", "vites", "yakit"], drop_first=True)      # Bu aşamada kategorik ifadeleri sayısal sorgu verilerine dönüştürdü. Bu sayede her ifade 1 ve 0 gibi rakamlarla ifade edilebilir oldu.
# drop_first=True --> Bu seçenek eklendiğinde ilk verileri kaldırarak sütun sayısını azaltır. 0 sa vardır ise 1 se yoktur. Ama burada tek veri yeterlidir. 0 olan veri kaldırılır.
df.head(5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df["fiyat"]
x = df.drop("fiyat", axis=1)            # Tüm verilerden "fiyat" ifadesinin çıkartılmış halidir. axis=1 ise fiyat sütununu hariç tutması için sütun olduğunu belirtir.
```

```python
lm = LinearRegression()
model = lm.fit(x, y)
```

```python
model.predict([[2017, 30000, 1.6, 110, 1, 2600, 0, 1]])     # Bağımsız değişkenler atandı. Bu değişkenlere göre fiyat tahmini yapıldı.
```

```python
model.score(x, y)           # Başarı oranı
```
